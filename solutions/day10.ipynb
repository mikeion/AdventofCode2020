## Day 10: Adapter Array

Patched into the aircraft's data port, you discover weather forecasts of a massive tropical storm. Before you can figure out 
whether it will impact your vacation plans, however, your device suddenly turns off!

Its battery is dead.

You'll need to plug it in. There's only one problem: the charging outlet near your seat produces the wrong number of jolts. 
Always prepared, you make a list of all of the joltage adapters in your bag.

Each of your joltage adapters is rated for a specific output joltage (your puzzle input). Any given adapter can take an input 
1, 2, or 3 jolts lower than its rating and still produce its rated output joltage.

In addition, your device has a built-in joltage adapter rated for 3 jolts higher than the highest-rated adapter in your bag. 
(If your adapter list were 3, 9, and 6, your device's built-in adapter would be rated for 12 jolts.)

Treat the charging outlet near your seat as having an effective joltage rating of 0.

Since you have some time to kill, you might as well test all of your adapters. Wouldn't want to get to your resort and realize 
you can't even charge your device!

If you use every adapter in your bag at once, what is the distribution of joltage differences between the charging outlet,

````
def count_differences(d):
    l = d.copy()
    l.insert(0,0)
    l.append(max(l)+3)
    diff_1 = 0
    diff_3 = 0
     # Taking into account the difference of the adapter at the end
    for i in range(len(l)-1):
        if l[i+1]-l[i] == 1:
            diff_1 +=1
        elif l[i+1]-l[i] == 3:
            diff_3 +=1
    return diff_1 * diff_3
````

You glance back down at your bag and try to remember why you brought so many adapters; there must be more than a trillion

Q: **What is the total number of distinct ways you can arrange the adapters to connect the charging outlet to your device?**

My idea here is that as we go through the list of adapters, we want to keep track of the paths they go down.

I'm thinking that we can set up a dictionary with an initial key, value pair of (0,1) given that there with always be a path connecting the charger to the
first adapter in our bag.

More generally, when we get to any particular adapter in the bag, we want to make sure we include all the paths that are within 1-, 2- and 3-jolt differences.
I'm thinking if we continue to collect these in the dictionary, then the number of paths at any given vertice $v$ will be 

$$ |paths(v)| = |paths(v-3)|+ |paths(v-2)| + |paths(v-1)|$$

To finish our function, we will just return the number of paths at the final number in the path (our adapter of our device).

Note: I need to use dictionary's get method, because keys will often be missing (see this stack exchange 
[post](https://stackoverflow.com/questions/11041405/why-dict-getkey-instead-of-dictkey))

````
def number_of_paths(data):
    d = {}
    d[0] = 1
    for number in data:
        d[number] = d.get(number-3, 0) + d.get(number-2, 0) + d.get(number-1, 0)
    print(d[max(d)])
````

````
number_of_paths(data)
````
